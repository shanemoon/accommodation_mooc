### Training

```python
import word2vec
```

Note that this will take a long time depending on the parameters

```python
word2vec.word2vec('data/text8', 'data/text8.bin', size=300, verbose=True)
```

That generated a `text8.bin` file containing the word vectors in a binary format.

### Predictions

```python
import word2vec
```

We import the binary file created above

```python
model = word2vec.load('data/text8.bin')
```

Words similar to socks:

```python
model.cosine('socks')
```

Can ask for multiple words at a time

```python
model.cosine(['black', 'shoe'], n=5)
```

Can do some analogies such as: `king - man + woman = queen` 

```python
model.analogy(pos=['king', 'woman'], neg=['man'], n=35)
```

Note that queen is not the first option on this case because of the parameters chosen, using a bigger vector size will give better results.

You can see the whole vocabulary

```python
model.vocab
```

Or get the vector for a word

```python
model['bird']
```
